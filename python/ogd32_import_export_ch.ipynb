```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# import and dispaly data
df_data = pd.read_csv('../data/ogd32_elektrizitaetbilanz_jahreswerte.csv', sep=',', encoding='utf-16')
df_data.head(32)
```

```python
# filter data
df_data = df_data[df_data['Jahr'] > 1989]
```

```python
# group data
df_import_export_ch = df_data[['Jahr', 'Einfuhr_GWh', 'Ausfuhr_GWh']].copy()

display(df_import_export_ch)
```

```python
# rename collumns
df_import_export_ch.rename(columns={'Jahr': 'Year'}, inplace=True)
df_import_export_ch.rename(columns={'Einfuhr_GWh': 'Import'}, inplace=True)
df_import_export_ch.rename(columns={'Ausfuhr_GWh': 'Export'}, inplace=True)

display(df_import_export_ch)
```

```python
# write json file
highcharts_series = []
for column in df_import_export_ch.columns:
    series_data = {
        'name': column,
        'data': df_import_export_ch[column].fillna(0).tolist()  # Replace NaN with 0 for Highcharts
    }
    highcharts_series.append(series_data) 

with open('../src/data/import-export-ch.json', 'w') as json_file:
    json.dump(highcharts_series, json_file, indent=2)
```
