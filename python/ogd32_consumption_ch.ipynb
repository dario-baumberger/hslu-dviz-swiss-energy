```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
#load data
df_data = pd.read_csv('../data/ogd32_elektrizitaetbilanz_jahreswerte.csv', sep=',', encoding='utf-16')
df_data = df_data[df_data['Jahr'] > 1989]
df_data.head()
```

```python
# filter columns
df_consumption  = df_data[['Jahr', 'Einfuhr_GWh', 'Ausfuhr_GWh', 'Landesverbrauch_GWh', 'Endverbrauch_GWh', 'Erzeugung_netto_GWh']].copy()
df_consumption.head()
```

```python
#Â rename columns

df_consumption.rename(columns={'Jahr': 'Year'}, inplace=True)
df_consumption.rename(columns={'Einfuhr_GWh': 'Import'}, inplace=True)
df_consumption.rename(columns={'Ausfuhr_GWh': 'Export'}, inplace=True)
df_consumption.rename(columns={'Landesverbrauch_GWh': 'Country consumption'}, inplace=True)
df_consumption.rename(columns={'Endverbrauch_GWh': 'End Consumption'}, inplace=True)
df_consumption.rename(columns={'Erzeugung_netto_GWh': 'Production Netto'}, inplace=True)

df_consumption.head()
```

```python
# write to json
highcharts_series = []
for column in df_consumption.columns:
    series_data = {
        'name': column,
        'data': df_consumption[column].fillna(0).tolist()  # Replace NaN with 0 for Highcharts
    }
    highcharts_series.append(series_data) 
    

with open('../src/data/consumption.json', 'w') as json_file:
    json.dump(highcharts_series, json_file, indent=2)
```
