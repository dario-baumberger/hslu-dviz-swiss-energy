```python
import pandas as pd
import matplotlib.pyplot as plt
import json

df_data = pd.read_excel('../data/Consumption_CH.xlsx', skiprows=8, usecols= 'A:B', names=['Year', 'Consumption'], nrows= 114)
display(df_data)
```

```python
highcharts_series = []
for column in df_data.columns:
    series_data = {
        'name': column,
        'data': df_data[column].fillna(0).tolist()  # Replace NaN with 0 for Highcharts
    }
    highcharts_series.append(series_data) 
    

with open('../public/data/consumption_CH.json', 'w') as json_file:
    json.dump(highcharts_series, json_file, indent=2)
```
